```python
# 기업 데이터

# 코드 과정
# Grid Search(2~10의 클러스터)를 통해 최적 클러스터 수 계산:

# Silhouette Score: 값이 가장 높은 클러스터 수 선택.
# Davies-Bouldin Score: 값이 가장 낮은 클러스터 수 선택.
# K-Means 최적화:

# 두 기준 중 하나를 선택하여 최적의 클러스터 수로 K-Means 수행.
# 여기서는 Silhouette Score 기준으로 클러스터 수를 설정.
# 클러스터 분석:
# 각 클러스터의 평균값을 확인하여 특성을 분석.
# PCA 시각화:
# 최적 클러스터 수로 군집화된 데이터를 2D로 축소하여 시각화.


import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
import numpy as np 

# 데이터 로드
data = pd.read_csv("C:/Users/USER/Downloads/DataAnalyst.csv/DataAnalyst.csv")

# 필요한 열 선택
df = data[['Salary Estimate', 'Rating', 'Size', 'Type of ownership', 'Industry', 'Revenue']]

# Salary Estimate 전처리
def preprocess_salary(salary):
    try:
        if isinstance(salary, str):
            salary = salary.split('(')[0]  # "(Glassdoor est.)" 제거
            salary = salary.replace('$', '').replace('K', '').replace(',', '')  # '$', 'K', ',' 제거
            if '-' in salary:
                min_salary, max_salary = salary.split('-')
                min_salary = int(min_salary.strip()) if min_salary.strip().isdigit() else None
                max_salary = int(max_salary.strip()) if max_salary.strip().isdigit() else None
                if min_salary is not None and max_salary is not None:
                    return (min_salary + max_salary) / 2  # 평균값 반환
        return None  # 변환 불가능한 경우
    except Exception as e:
        print(f"Error processing salary: {salary}, Error: {e}")
        return None

df['Salary Estimate'] = df['Salary Estimate'].apply(preprocess_salary)

# 결측값 처리
df = df.dropna()  # 결측값 제거

# 범주형을 수치형으로, 라벨인코딩
label_encoders = {}
for col in ['Size', 'Type of ownership', 'Industry', 'Revenue']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# 데이터 정규화 (단위를 맞춤)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Grid Search를 사용한 클러스터 수 탐색
# 최적의 클러스터 수를 찾기 위한 범위
cluster_range = range(2, 11)

# 평가 점수 저장 리스트
silhouette_scores = []
davies_bouldin_scores = []

# 클러스터 수별로 점수 계산
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_data)
    
    # Silhouette Score 계산
    silhouette_avg = silhouette_score(scaled_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    
    # Davies-Bouldin Score 계산
    davies_bouldin_avg = davies_bouldin_score(scaled_data, cluster_labels)
    davies_bouldin_scores.append(davies_bouldin_avg)
    
    print(f"Clusters: {n_clusters}, Silhouette Score: {silhouette_avg:.4f}, Davies-Bouldin Score: {davies_bouldin_avg:.4f}")

# 최적의 클러스터 수 찾기
optimal_clusters_silhouette = cluster_range[np.argmax(silhouette_scores)]
optimal_clusters_davies = cluster_range[np.argmin(davies_bouldin_scores)]

print(f"\nOptimal Clusters based on Silhouette Score: {optimal_clusters_silhouette}")
print(f"Optimal Clusters based on Davies-Bouldin Score: {optimal_clusters_davies}")

# Silhouette Score 기준 최적의 클러스터 수로 K-Means 수행
kmeans = KMeans(n_clusters=optimal_clusters_silhouette, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)

# 클러스터 결과 분석
print("\nCluster Analysis (Based on Silhouette Score):")
print(df.groupby('Cluster').mean())

# 결과 시각화 (2차원 PCA 사용)
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.7)
plt.colorbar()
plt.title(f"K-Means Clustering Results (Optimal Clusters: {optimal_clusters_silhouette})")
plt.show()

```
